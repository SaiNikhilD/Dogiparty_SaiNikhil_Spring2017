To Analyse which share of the technology market has maximum returns

```python
#Importing all the necessary libarires for Analysis 1
import pandas as pd
import pandas_datareader  as web   
import datetime
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rc('figure', figsize=(15, 10))

from matplotlib.dates import DateFormatter, WeekdayLocator,\
    DayLocator, MONDAY,date2num
from matplotlib.finance import candlestick_ohlc
import matplotlib.pyplot as plt  
```

To look at the stock market for only 2017 years records

```python
# Setting start and end date to be used later
start = datetime.datetime(2017,1,1)
end = datetime.date.today()
```

#### Loading data for the required securities/shares into dataframes

```python
microsoft = pd.read_csv('/data/MSFT.csv', index_col='Date', parse_dates=True)
microsoft = microsoft.reset_index()
mask = (microsoft['Date'] > start) & (microsoft['Date'] <= end)
microsoft = microsoft.loc[mask]
microsoft = microsoft.set_index(['Date'])
microsoft.head()
```

```python
apple = pd.read_csv('/data/AAPL.csv', index_col='Date', parse_dates=True)
apple = apple.reset_index()
mask = (apple['Date'] > start) & (apple['Date'] <= end)
apple = apple.loc[mask]
apple = apple.set_index(['Date'])
apple.head()
```

```python
google = pd.read_csv('/data/GOOGL.csv', index_col='Date', parse_dates=True)
google = google.reset_index()
mask = (google['Date'] > start) & (google['Date'] <= end)
google = google.loc[mask]
google = google.set_index(['Date'])
google.head()
```

```python
yahoo = pd.read_csv('/data/YHOO.csv', index_col='Date', parse_dates=True)
yahoo = yahoo.reset_index()
mask = (yahoo['Date'] > start) & (yahoo['Date'] <= end)
yahoo = yahoo.loc[mask]
yahoo = yahoo.set_index(['Date'])
yahoo.head()
```

```python
cisco = pd.read_csv('/data/CSCO.csv', index_col='Date', parse_dates=True)
cisco = cisco.reset_index()
mask = (cisco['Date'] > start) & (cisco['Date'] <= end)
cisco = cisco.loc[mask]
cisco = cisco.set_index(['Date'])
cisco.head()
```

#### Stock return is plotted

```python
#join method to combine dataframes is used
techstocks = pd.DataFrame({"AAPL": apple["Adj. Close"],
                      "MSFT": microsoft["Adj. Close"],
                      "GOOG": google["Adj. Close"],
                      "YHOO": yahoo["Adj. Close"],
                      "CISCO": cisco["Adj. Close"],
                      }) 

#Using lambda function 
stock_return = techstocks.apply(lambda x: x / x[0])
stock_return.head()

stock_return.plot(grid = True).axhline(y = 1, color = "black", lw = 2)

plt.ylabel('Stock Return')

plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
plt.title("STOCK RETURN PLOT")
plt.savefig('stockreturn.png', dpi = 300)
plt.show()

```

The famous japanese candle stick has been plotted using the below function 

```python
def nikhil_plotter(share, stick = "day", otherseries = None):
    
    # for marking special days
    mondays = WeekdayLocator(MONDAY)        
    alldays = DayLocator()              
    dayFormatter = DateFormatter('%d')      
 
    #df containing the required details
    reqdat = share.loc[:,["Open", "High", "Low", "Close"]]
    if (type(stick) == str):
        if stick == "day":
            plotshare = reqdat
            stick = 1 # Used for plotting
    
    else:
        raise ValueError('Valid inputs to argument "stick" include the strings "day", "week", "month", "year", or a positive integer')
 
 
    # Set plot parameters, including the axis object ax used for plotting
    fig, ax = plt.subplots()
    fig.subplots_adjust(bottom=0.2)
    if plotshare.index[-1] - plotshare.index[0] < pd.Timedelta('360 days'):
        weekFormatter = DateFormatter('%b %d')  
        ax.xaxis.set_major_locator(mondays)
        ax.xaxis.set_minor_locator(alldays)
    else:
        weekFormatter = DateFormatter('%b %d, %Y')
        ax.xaxis.set_major_formatter(weekFormatter)
 
    ax.grid(True)
 
    # candelstick chart
    candlestick_ohlc(ax, list(zip(list(date2num(plotshare.index.tolist())), plotshare["Open"].tolist(), plotshare["High"].tolist(),
                      plotshare["Low"].tolist(), plotshare["Close"].tolist())),
                      colorup = "black", colordown = "red", width = stick * .4)
 
 
    ax.xaxis_date()
    ax.autoscale_view()
    plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')
 
    plt.title("CandleStick PLOT")
    plt.ylabel('Price')
    plt.xlabel('Dates')
    
    
    plt.legend()
    plt.show() 
```

#### Plot the candle stick graph for the top 3 firms of technology analyzed

```python
nikhil_plotter(cisco)
```

```python
nikhil_plotter(apple)
```

```python
nikhil_plotter(yahoo)
```
