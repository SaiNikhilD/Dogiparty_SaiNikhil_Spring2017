- To calculate the moving average and plot against the regular adjusted closing
- The percentage daily return is also calculated and depicted in a plot

Importing all the necessary libarires for Analysis 2

```python
import datetime
import pandas as pd
import pandas_datareader 
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rc('figure', figsize=(15, 15))
```

```python
start = datetime.datetime(2017,1,1)
end = datetime.date.today()
```

Plotted Moving Average of the top 3 analysed shares

What is a 'Moving Average - MA'
A widely used indicator in technical analysis that helps smooth out price action by filtering out the “noise”
from random price fluctuations. 
A moving average (MA) is a trend-following or lagging indicator because it is
based on past prices.

```python
# top 3 analyzed shares
shares = {'GOOGL','CSCO','MSFT'}

# for each mechanism used to avoid redundant code
for each in shares:
      
    df = pd.read_csv('/data/'+each+'.csv', index_col='Date', parse_dates=True)
   
    df = df.reset_index()
    #masking dataframe to get details of a 2017 year
    mask = (df['Date'] > start) & (df['Date'] <= end)
    df = df.loc[mask]

    df = df.set_index(['Date'])
    #selecting only required columns
    close_px = df['Adj. Close']
    
    #moving average is being calculated using mean
    mavg = close_px.rolling(window=5,center=False).mean()
    
    # plot
    close_px.pct_change().head()
    close_px.plot(label=each)
    mavg.plot(label='mavg')
    plt.legend()
    plt.ylabel('ADJUSTED CLOSING')
    plt.title("MOVING AVERAGE")
    plt.show()
    
    #percentage daily return is calcuated and plotted
    df['Adj. Close'].pct_change().plot(figsize=(20,10),legend=True,linestyle='--',marker='o')
    plt.ylabel('PERCENTAGE')
    plt.title("DAILY RETURN PERCENTAGE PLOT")
    plt.show()
    
```
