# Analysis 1

To print the details of the TV series

```python
import json,codecs,glob


#relative file path
jsonfiles=glob.glob('Data/*.json')  

for inputfile in jsonfiles:     
    
    data_file = codecs.open(inputfile, 'r', encoding='utf8',errors='ignore')
    data = json.load(data_file)
    docs = data["response"]["docs"]
    
    for each in docs:
        print (each['type_of_material'])
        print (each['word_count'])
        print (each['section_name'])
        print (each['lead_paragraph'])
    
   
```
