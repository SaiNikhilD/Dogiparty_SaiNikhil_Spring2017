# Analysis 1_1

To print the article topics trend of the January 2017 TV series

```python
import json,codecs,glob
sec={}
import matplotlib.pyplot as mp

#relative file path
jsonfiles=glob.glob('Data/*.json')  

for inputfile in jsonfiles:     
    
    data_file = codecs.open(inputfile, 'r', encoding='utf8',errors='ignore')
    data = json.load(data_file)
    docs = data["response"]["docs"]
    
    for each in docs:
                
        if each['section_name'] not in sec:
            sec[each['section_name']]=1
        else:
            sec[each['section_name']]+=1
    
```

```python
topic=[]
topic_count=[]
for a in sec:
    topic.append(a)
    topic_count.append(sec[a])
```

```python
mp.style.use('ggplot')
mp.style.use('seaborn-dark-palette')
mp.style.use('seaborn-poster')


mp.bar(range(len(topic)),topic_count,color='blue',align='center')
mp.xticks(range(len(topic)),topic,rotation=90)
mp.title("'Topic for January'17")
mp.ylabel('Count')
mp.xlabel('Topic')

mp.show()
```
